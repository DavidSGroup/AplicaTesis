```python
import json
import numpy as np
import tensorflow.keras as keras
#48erk,
OUTPUT_UNITS = 48
NUM_UNITS = [256]
LOSS = "sparse_categorical_crossentropy"
LEARNING_RATE = 0.001
EPOCHS =50       #50
BATCH_SIZE = 64
ROUTE_MODEL= "C:/Users/{}/Documents/AplicaTesis/3MODELO/cuecapura1.h5".format(logger)
```

```python

def build_model(output_units, num_units, loss, learning_rate):

    # create the model architecture
    input = keras.layers.Input(shape=(None, output_units))
    #input = keras.layers.Input(shape=(timesteps, input_features))
    x = keras.layers.LSTM(num_units[0])(input)
    x = keras.layers.Dropout(0.2)(x)

    output = keras.layers.Dense(output_units, activation="softmax")(x)

    model = keras.Model(input, output)

    # compile model
    model.compile(loss=loss,
                  optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
                  metrics=["accuracy"])

    model.summary()

    return model
```

```python

def develop_model(output_units=OUTPUT_UNITS, num_units=NUM_UNITS, loss=LOSS, learning_rate=LEARNING_RATE):
  
    # generate the training sequences
    
    inputs, targets = gtraining_sequences(SQ_LENGTH)

    # build the network
    model = build_model(output_units, num_units, loss, learning_rate)

    # train the model
    model.fit(inputs, targets, epochs=EPOCHS, batch_size=BATCH_SIZE)
    #perdidas y precision
    resultados = model.evaluate(inputs, targets, verbose=0)

    # Obtener la pérdida y precisión
    perdida = resultados[0]
    precision = resultados[1]

    # Imprimir la pérdida y precisión en porcentaje
    print(f'Pérdida: {perdida:.2f}')
    print(f'Precisión: {precision * 100:.2f}%')

    # save the model
    model.save(ROUTE_MODEL)
```

```python
if __name__ == "__main__":
    develop_model()
```
